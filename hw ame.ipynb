```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Parameters
r = 50
img_h = 4 * r
img_w = 4 * r
img = np.zeros((img_h, img_w), dtype=np.uint8)

# Center
cx, cy = img_w // 2, img_h // 2

cv2.rectangle(img, (cx - r, cy - r ), (cx + r, cy + r ), 255, -1)

# Top semicircle
cv2.ellipse(img, (cx, cy + r), (r, r), 0, 0, 180, 255, -1)

# Bottom semicircle
cv2.ellipse(img, (cx, cy - r), (r, r), 0, 180, 360, 255, -1)

# Left triangle
pts_left = np.array([
    [cx - r, cy - r],
    [cx - 2*r, cy],
    [cx - r, cy + r]
], np.int32)
cv2.fillPoly(img, [pts_left], 255)

# Right triangle
pts_right = np.array([
    [cx + r, cy - r],
    [cx + 2*r, cy],
    [cx + r, cy + r]
], np.int32)
cv2.fillPoly(img, [pts_right], 255)

cv2.rectangle(img,
              (cx - r // 2, cy - r // 2),  # Top-left corner
              (cx + r // 2, cy + r // 2),  # Bottom-right corner
              0,                          # Black color
              -1)                         # Fill the square

plt.figure(figsize=(5, 5))
plt.title("Binary Image with Specified Shape")
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```
